El proceso de entrenamiento en machine learning es un proceso central para cualquier tarea de aprendizaje automático que busca la optimización sin caer en el sobreajuste, asegurando así que el modelo sea tanto preciso como generalizable. El modelo, mediante el algoritmo, aprende automáticamente de los datos sin intervención humana directa en los ajustes específicos de los parámetros. En este sentido, es muy importante regular el proceso de entrenamiento para asegurar que el modelo sea generalizable y no solo memorice los datos de entrenamiento. Este procedimiento se caracteriza por varios aspectos fundamentales:

1. **Ajuste de Parámetros**: el núcleo del entrenamiento de un modelo es el ajuste de sus parámetros internos, tales como los pesos en una red neuronal. El objetivo es minimizar una función de pérdida que mide la discrepancia entre las predicciones del modelo y las salidas reales en los datos de entrenamiento.


2. **Algoritmo de Aprendizaje**: para ajustar los parámetros, se utiliza un algoritmo de aprendizaje. Los más comunes incluyen el descenso de gradiente (y sus variantes) para problemas de regresión y clasificación, o algoritmos como el EM (Expectation-Maximization) para modelos más complejos. El algoritmo iterativamente propone cambios en los parámetros del modelo para mejorar el rendimiento del modelo según la función de pérdida.


3. **Iteraciones**: el proceso de entrenamiento es iterativo. En cada iteración, el algoritmo de aprendizaje ajusta los parámetros del modelo en respuesta a los datos. Este proceso continúa hasta que se alcanza un criterio de parada, como un número máximo de iteraciones, un cambio mínimo en la función de pérdida entre iteraciones sucesivas, o una señal de que el modelo ha comenzado a sobreajustarse.


4. **Optimización de la Función de Pérdida**: durante el entrenamiento, el enfoque está en optimizar la función de pérdida. Esta función varía según el tipo de tarea: puede ser la suma de cuadrados para regresión, la entropía cruzada para clasificación, entre otras. La optimización de esta función es lo que guía el proceso de aprendizaje para mejorar el modelo.


5. **Uso de los Datos de Entrenamiento**: Solo los datos de entrenamiento se usan durante esta fase. Estos datos deben representar de manera adecuada el universo de situaciones que el modelo enfrentará, para que pueda aprender patrones generalizables. Sin embargo, es vital que este conjunto de datos esté libre de errores y sea lo suficientemente diverso para evitar el sesgo y el sobreajuste.

<div style="background-color:#5DADE2"><b>Apartado 2</b></div>

Los modelos y técnicas de clasificación en aprendizaje supervisado son diversas y se utilizan para predecir la categoría o clasificación de instancias o datos de entrada. Estas técnicas varían en complejidad, tipo de modelo, y la forma en que tratan con los datos. A continuación, se enumeran algunas de los <u>principales modelos de clasificación</u> junto con una breve descripción de cada uno de ellos:

* **Regresión Logística**: A pesar de su nombre, es un modelo lineal para clasificación binaria que estima probabilidades utilizando una función logística. Es útil para casos donde la variable objetivo es binaria.


* **Árboles de Decisión**: Modelo que utiliza una estructura de árbol para tomar decisiones basadas en las características de los datos. Cada nodo del árbol representa una característica, cada rama un criterio de decisión, y cada hoja una clasificación.


* **Bosques Aleatorios (Random Forest)**: Un ensamble de múltiples árboles de decisión para mejorar la robustez y precisión de la clasificación. Utiliza técnicas de bagging para crear diferentes conjuntos de datos de entrenamiento para cada árbol, lo que ayuda a reducir el sobreajuste.


* **Máquinas de Vector Soporte (SVM)**: Busca el hiperplano que mejor separa las clases en el espacio de características. Es eficaz en espacios de alta dimensión y casos donde el número de dimensiones supera el número de muestras.


* **K-Vecinos más cercanos (K-NN)**: Clasifica las instancias basándose en la mayoría de votos de los k vecinos más cercanos en el espacio de características. Es un método no paramétrico que es simple pero efectivo para conjuntos de datos con muchas clases.


* **Naive Bayes**: Un clasificador probabilístico basado en el teorema de Bayes con la "naive" que significa suposición de independencia entre las variables de entrada. A pesar de su simplicidad, funciona bien en muchas situaciones reales.

Si la exactitud de las predicciones (por ejemplo, porcentaje de casos bien clasificados en un problema de clasificación) en el conjunto de test es (además de elevada) similar a la que se obtiene en el conjunto de entrenamiento, entonces el modelo entrenado generalizará bien para otros conjuntos de datos y puede darse por bueno. En otro caso, el modelo no ha entrenado bien y se nos podrían plantear dos circunstancias:

* Problema de **sobreajuste u overfitting**: por ejemplo, si la exactitud de las predicciones en el conjunto de entrenamiento es del $80\%$ y en el de test es del $25\%$ o del $60\%$, entonces el modelo no puede darse por válido. * Problema de **infraajuste o underfitting**: se da cuando un modelo es demasiado simple para capturar la estructura subyacente de los datos. Esto significa que el modelo no aprende suficientemente de los datos de entrenamiento, resultando en una incapacidad para hacer predicciones precisas tanto en los datos de entrenamiento como en los nuevos. La simplicidad del modelo puede deberse a no tener suficientes parámetros, a una selección inadecuada de características o a una falta de complejidad en el modelo elegido para captar las relaciones entre las características. <div style="background-color:#5DADE2"><b>Apartado 1</b></div>

El bootstrap es una técnica de remuestreo ampliamente utilizada en el entrenamiento de algoritmos de machine learning, especialmente en contextos donde es necesario estimar la variabilidad de los datos o mejorar la robustez del modelo. Consiste en generar múltiples muestras aleatorias con reemplazo a partir de una muestra original, lo que permite construir una distribución empírica de las estadísticas de interés o entrenar modelos en subconjuntos variados de los datos.

La técnica tiene como características principales su independencia de suposiciones estrictas sobre la distribución de los datos, su capacidad para trabajar con tamaños de muestra moderados y su enfoque intensivo en cálculos debido al remuestreo repetido. El bootstrap es especialmente valioso cuando la muestra original se considera representativa de la población, ya que permite capturar de forma efectiva la variabilidad inherente en los datos. 

Además, facilita la construcción de intervalos de confianza para parámetros clave del modelo, lo que mejora la interpretación de los resultados. No obstante, su eficacia puede disminuir si los datos originales presentan sesgos significativos o si el tamaño de la muestra es muy pequeño. A pesar de estas limitaciones, el bootstrap es una herramienta poderosa y versátil en el machine learning moderno, favoreciendo tanto la mejora del rendimiento de los modelos como la evaluación precisa de su incertidumbre.

<div style="background-color:#5DADE2"><b>Apartado 2</b></div>

La técnica de validación cruzada k-fold es un método estadístico usado para evaluar y mejorar la robustez de un modelo de machine learning, especialmente en términos de su capacidad para generalizar a nuevos datos que no se utilizaron durante el entrenamiento. Es una técnica esencial para evitar problemas como el sobreajuste y para asegurar que el modelo no esté sesgado por la partición particular de los datos en conjuntos de entrenamiento y prueba. Como sabemos, en la validación cruzada $k$-fold, el conjunto de datos se divide aleatoriamente en $k$ subconjuntos (o "folds") de aproximadamente igual tamaño. El modelo se entrena y evalúa $k$ veces, cambiando el fold que se usa como conjunto de prueba en cada iteración y utilizando los $k−1$ folds restantes como conjunto de entrenamiento. En resumen, el paso por paso para poder aplicar dicha técnica sería el siguiente:

1. **Dividir el Conjunto de Datos**: El conjunto total de datos se divide en $k$ subconjuntos o folds. Estos folds se crean para que sean mutuamente exclusivos y colectivamente exhaustivos, es decir, cada registro del conjunto de datos debe estar exactamente en uno de los $k$ folds.


2. **Iteraciones de Entrenamiento y Evaluación**: Para cada una de las $k$ iteraciones se selecciona un fold diferente como el conjunto de datos de prueba. Los $k−1$ folds restantes se utilizan como conjunto de datos de entrenamiento. Se entrena el modelo en el conjunto de entrenamiento y se evalúa el modelo en el conjunto de prueba. 


3. **Promedio de Resultados**: Al final de las $k$ iteraciones, se calcula el promedio de la métrica de evaluación registrada en cada iteración. Esto proporciona una medida única que es menos sensible a la partición de los datos en entrenamiento y prueba, dando una mejor indicación del rendimiento esperado del modelo en nuevos datos.

Por tanto, al utilizar diferentes conjuntos de datos de prueba en cada iteración, el modelo no puede simplemente memorizar los datos, lo que ayuda a prevenir el sobreajuste. Además, el rendimiento medio obtenido es generalmente más estable y fiable que el obtenido a partir de una única división de entrenamiento/prueba, porque reduce la variabilidad que podría resultar de una partición aleatoria infortunada de los datos.

<div style="background-color:#5DADE2"><b>Apartado 3</b></div>

Las métricas que tenemos que considerar para realizar la evaluación del modelo en un problema de clasificación son las siguientes:

- **Matriz de confusión**: que nos muestra una tabla con los verdaderos positivos (VP), los verdaderos negativos (VN), los falsos positivos (FP) y los falsos negativos (FN).


- **Precisión**: mide la proporción de identificaciones positivas que fueron correctas. Se calcula como VP / (VP + FP), lo que indica qué tan bien el modelo evita falsos positivos.


- **Recall (Sensibilidad)**: evalúa la capacidad del modelo para identificar correctamente todas las observaciones positivas. Se calcula como VP / (VP + FN), mostrando cuán bien el modelo detecta casos positivos.


- **Especificidad**: mide la proporción de verdaderos negativos identificados correctamente. Es el complemento de la sensibilidad. Se calcula como VN / (VN + FN), indicando cuán bien el modelo detecta casos negativos.


- **Puntuación F1**: combina la precisión y el recall en una sola métrica mediante su media armónica. Calculada como $2 * (Precisión * Recall) / (Precisión + Recall)$, esta métrica es útil cuando se necesita un equilibrio entre precisión y recall. Su rango de valores oscila entre $[0, 1]$. Es decir, el equilibrio entre la captura de positivos y la precisión con la que los identifica.


- **AUC-ROC (Área bajo la Curva del Operador Receptor)**: esta métrica evalúa la capacidad del modelo para discriminar entre clases. Un área de $1$ representa un modelo perfecto, mientras que un área de $0.5$ indica un rendimiento no mejor que el azar.

En este paso, entrenamos el modelo de regresión logística (`glm`) utilizando todas las variables consideradas para clasificar a los clientes entre pagadores o morosos. La configuración específica del entrenamiento se tiene que definir previamente a través de la función **`trainControl()`**. En primer lugar, utilizaremos la técnica de validación cruzada de $10$ pliegues. Para ello, el conjunto de entrenamiento se dividirá en $10$ bloques, entrenando el modelo en $9$ de ellas y validando en la restante, rotando hasta que todas las partes hayan sido utilizadas para la validación.

```r
# Configuramos el control de entrenamiento para usar validación cruzada de 10 pliegues, lo que ayuda a evitar 
# el sobreajuste y proporciona una estimación más robusta del rendimiento del modelo
trainControl = caret::trainControl(method = "repeatedcv", number = 10, repeats = 10)
```
