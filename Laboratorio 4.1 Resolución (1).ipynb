En este caso, la variable puede no ser relevante para el modelo, y su inclusin podra no aportar mucha informacin o mejora, aunque tenga significatividad estadstica.

#### 4.9 Cmo extraer el coeficiente de determinacin $R^2$ de McFadden?

En el modelo de regresin logstica no tiene sentido calcular el coeficiente de determinacin lineal, $R^2$, ya que estamos modelizando dicha relacin a travs de la ecuacin logstica, pero existen varias alternativas equivalentes para hacerse una idea de la variabilidad de la respuesta explicada por el modelo. Las tres ms populares son el pseudo $R^2$ de McFadden (que ser el que utilicemos nosotros), el $R^2$ de Cox y Snell (que por construccin no puede alcanzar el $1$) y el $R^2$ de Nagelkerke (una correccin del de Cox y Snell). Vamos a extraer el coeficiente de determinacin $R^2$ de McFadden a travs de una forma nicamente:

+ Clculo manual del coeficiente de determinacin de McFadden a partir de la siguiente ecuacin formal:

\begin{align*}
R^{2}=1 - \frac{Log-Likelihood_1}{Log-Likelihood_0}
\end{align*}

siendo $Log-Likelihood_1$ el logaritmo de la razn de verosimilitud del modelo de regresin logstico con todas las variables y $Log-Likelihood_0$ el logaritmo de la razn de verosimilitud del modelo de regresin logstico constante.

```r
# Estimamos el modelo de regresin logstica constante
null = glm(low ~ 1, family = "binomial", data = wdata)
```

```r
# Estimamos el coeficiente de determinacin R2 de McFadden
1 - logLik(modelo_logi) / logLik(null)
```

**<u>Interpretacin resultados</u>.-** Los valores pueden variar entre $0$ (indicando que las variables independientes no son buenas explicando la probabilidad de la variable respuesta) y $1$ (indicando que el modelo explica bien la probabilidad de la variable respuesta). En este caso, obtenemos un valor $R^2$ de McFadden igual a $0.02548803$ lo que indica que considerando solamente la variable `lwt` no conseguimos mucha capacidad explicativa de la probabilidad de que el peso de un recin nacido sea inferior a $2.5$kg. Luego, si bien, el modelo mejora al cincorporar dicha variable, frente al modelo nulo o constante, sera recomendable incorporar ms variables explicativas para mejorar los resultados.

#### 4.9 Cmo extraer el coeficiente de determinacin $R^2$ de McFadden?

Recordar que en el modelo de regresin logstica no tiene sentido calcular el coeficiente de determinacin lineal, $R^2$, ya que estamos modelizando dicha relacin a travs de la ecuacin logstica, pero existen varias alternativas equivalentes para hacerse una idea de la variabilidad de la respuesta explicada por el modelo. Las tres ms populares son el pseudo $R^2$ de McFadden (que ser el que utilicemos nosotros), el $R^2$ de Cox y Snell (que por construccin no puede alcanzar el $1$) y el $R^2$ de Nagelkerke (una correccin del de Cox y Snell). Vamos a extraer el coeficiente de determinacin $R^2$ de McFadden a travs de una forma nicamente:

+ Clculo manual del coeficiente de determinacin de McFadden a partir de la siguiente ecuacin formal:

\begin{align*}
R^{2}=1 - \frac{Log-Likelihood_1}{Log-Likelihood_0}
\end{align*}

siendo $Log-Likelihood_1$ el logaritmo de la razn de verosimilitud del modelo de regresin logstico con todas las variables y $Log-Likelihood_0$ el logaritmo de la razn de verosimilitud del modelo de regresin logstico constante.

```r
# Estimamos el modelo de regresin logstica constante
null = glm(Survived ~ 1, family = "binomial", data = wdata)
```

```r
# Estimamos el coeficiente de determinacin R2 de McFadden
1 - logLik(modelo_logi) / logLik(null)
```
