Un problema de **regresión** se diferencia de uno de **clasificación** principalmente por la forma en que se abordan los resultados y el tipo de variable objetivo $y$ considerada. En un problema de regresión, dicha variable es de tipo continuo, como el precio de una vivienda o la temperatura, lo que significa que puede tomar cualquier valor dentro de un rango numérico. En su lugar, en un problema de clasificación, la variable objetivo es categórica, lo que implica que pertenece a un conjunto específico de clases, como determinar si un correo electrónico es "spam" o "no spam", o identificar un animal en una imagen como "gato", "perro" o "pájaro".

Además, los métodos de evaluación son distintos: en regresión se usan métricas como el coeficiente de determinación ($R^2$), criterio de información bayesiano (BIC) o de Akaike (AIC), mientras que en clasificación se aplican medidas como la precisión, la exactitud, el recall o el F1-score. Los algoritmos o modelos utilizados en cada caso también reflejan estas diferencias, siendo ejemplos comunes de regresión el modelo de regresión lineal, la regresión polinómica o la regresión no-lineal con funciones step, mientras que en problemas de clasificación destacan los modelos de regresión logística, los árboles de decisión o los k-vecinos cercanos.

Durante esta etapa, el modelo ajusta sus parámetros internos utilizando técnicas matemáticas y estadísticas para minimizar la diferencia entre los valores ajustados y los valores reales.


* **Mejora del modelo.** Si el modelo no alcanza el rendimiento deseado, se implementan estrategias para mejorarlo. Por ejemplo, se obtienen datos adicionales o más diversos para mejorar el aprendizaje del modelo. O bien, se puede experimentar con modelos más complejos o con diferentes arquitecturas, como pasar de árboles de decisión a random forests.

<div style="background-color:#5DADE2"><b>Apartados 2-3 </b></div>

Los problemas de clasificación supervisada se dividen en tres tipos principales: binarios, multiclase y multietiqueta. Cada uno tiene características específicas que determinan cómo se abordan y aplican en diferentes contextos.

* El problema de **clasificación binaria** es el más simple de los tres. En este caso, el objetivo es asignar cada instancia de datos a una de dos clases posibles, típicamente etiquetadas como "positivo" y "negativo" (por ejemplo, "spam" y "no spam"). Este tipo de problema es útil en escenarios donde solo hay dos resultados posibles. Las predicciones suelen representarse como $0$ y $1$, y la función de pérdida comúnmente utilizada es la entropía cruzada o log-loss. Entre las métricas de evaluación más relevantes están la exactitud, precisión, recall, F1-score, y la curva ROC con su AUC, especialmente cuando las clases están desbalanceadas. Algunos ejemplos incluyen el diagnóstico de enfermedades ("sano" o "enfermo"), la detección de fraudes en transacciones financieras y la clasificación de correos electrónicos.


* En el problema de **clasificación multiclase**, cada instancia debe asignarse a una de varias clases posibles, con más de dos opciones mutuamente excluyentes. Por ejemplo, clasificar una imagen como "gato", "perro" o "pájaro". Aquí, el modelo predice una probabilidad para cada clase y selecciona la de mayor valor como resultado. Este tipo de problema utiliza funciones de pérdida como la entropía cruzada categórica y métricas de evaluación como la exactitud y la matriz de confusión para analizar el desempeño en cada clase. También se emplean métricas macro y micro promedio para problemas con clases desbalanceadas. Casos prácticos incluyen el reconocimiento de objetos en imágenes, la categorización de correos en "promociones", "social" o "trabajo", y la clasificación de enfermedades en diferentes tipos, por ejemplo.


* Por último, el problema de **clasificación multietiqueta** es más complejo, ya que una instancia puede pertenecer a varias clases simultáneamente. Las etiquetas no son mutuamente excluyentes, y cada clase genera una salida binaria independiente, lo que resulta en un vector de etiquetas. Este enfoque es útil en contextos donde una instancia puede tener múltiples características, como una imagen que se etiqueta simultáneamente como "montaña", "nieve" y "cielo". Las métricas de evaluación incluyen la exactitud basada en conjuntos (proporción de instancias etiquetadas correctamente con todas sus etiquetas), el Hamming loss (proporción de etiquetas mal clasificadas), y medidas de precisión, recall y F1-score promediadas por etiquetas. Aplicaciones comunes incluyen sistemas de recomendación de productos, clasificación de documentos con múltiples categorías, y anotación de imágenes con múltiples objetos o contextos. Este tipo de problemas no se verán de momento.

<div style="background-color:#5DADE2"><b>Apartado 3: Estimación de un modelo de regresión logístico simple entre las variables seleccionadas</b></div>

El Modelo de Regresión Logístico Simple para explicar la probabilidad de que el peso al nacer del bebé fuera inferior a $2.5$kg o no en función del peso de la madre en la última menstruación sería el siguiente: 

\begin{align*}
P\left ( Low \right ) = \frac{1}{1+e^-{\left ( \beta_0 + \beta_1 * Lwt \right )}}
\end{align*}

donde $P\left ( Low \right )$ es la probabilidad de que el peso al nacer del bebé fuera inferior a $2.5$kg o no, $e$ es la función exponencial y el resto de coeficientes son análogos a los ya vistos. Por tanto, el objetivo de la regresión logística NO es, como en regresión lineal o no-lineal, explicar el valor de la variable dependiente a partir de una o varias variables independientes, sino que queremos explicar la probabilidad de que ocurra (no el valor) la variable dependiente conocidos los valores de las variables independientes. Pasemos a estimar dicho modelo utilizando la siguiente función:

```r
# Estimación del modelo de regresión logística simple
modelo_logi = glm(low ~ lwt, family="binomial", data = wdata)  
modelo_logi
```

**<u>Interpretación resultados</u>.-** La interpretación de los coeficientes de regresión logística es un poco más complicada que la de la lineal, ya que representan el cambio en el logaritmo de la **odds ratio (OR)** de que el evento ocurra (la categoría $1$ de la variable dependiente, que el peso al nacer del bebé sea menor que $2.5$kg) en respuesta a un cambio unitario en la variable independiente del coeficiente (en el peso de la madre en la última menstruación). Repasemos la teoría.

Para calcular la OR de una variable, tendremos que aplicar la función exponencial (el anti-logaritmo natural) al coeficiente de regresión de esa variable (lo que solo tiene sentido si es estadísticamente significativo como ocurre en este caso) pudiendo obtener los siguientes resultados:

* Un <u>coeficiente de regresión positivo</u> (equivale a una OR mayor de 1) indicando que un aumento en la variable indipendiente, aumenta la probabilidad de que se produzca el evento (peso del bebé menor que $2.5$kg).


* Aquellos <u>coeficientes muy pequeños (próximos a cero)</u> nos darán OR próximas a 1, el valor nulo de la OR. En general, una OR cercana a 1 sugiere que el efecto de dicha variable independiente en la probabilidad de que se produzca el evento (peso del bebé menor que $2.5$kg) es prácticamente despreciable. En este caso, la variable puede no ser relevante para el modelo, y su inclusión podría no aportar mucha información o mejora, aunque tenga significatividad estadística.


* Un <u>coeficiente de regresión negativo</u> (equivale a una OR menor de 1) indicando que un aumento en la variable independiente, disminuye la probabilidad de que se produzca el evento (peso del bebé menor que $2.5$kg).


En este caso, solamente $\beta_1$ es estadísticamente significativa siendo el p-valor $< 0.05$, rechazando la hipótesis nula de que $\beta_1 = 0$. El peso de la madre en la última menstruación, variable cuantitativa, tiene un coeficiente de regresión de $\beta_1=-0.01405826$. Si le aplicamos la función exponencial $(exp(-0.01405826))$ obtenemos una $OR = 0.98604009589$. Como podemos ver, estaríamos en el caso de una $OR < 1$ sugieriendo que 
un aumento en la variable independiente, el peso de la madre, disminuiría la probabilidad de que el peso del bebé al nacer fuera menor que $2.5$kg. Según las leyes de la biología, se cumple que, en la mayoría de casos, conforme mayor sea el peso de la madre en la última menstruación mayor será el peso del recién nacido y, por tanto, menor probabilidad a que su peso sema menor a $2.5$kg. Por otro lado, el coeficiente constante $\beta_0$ nos ha salido no significativo, luego si le aplicamos la función exponencial $(exp(0))$ obtenemos una $OR = 1$ que significa que la constante no aporta información al modelo, es irrelevante su inclusión.

#### 4.9 ¿Cómo extraer el coeficiente de determinación $R^2$ de McFadden?

En el modelo de regresión logística no tiene sentido calcular el coeficiente de determinación lineal, $R^2$, ya que estamos modelizando dicha relación a través de la ecuación logística, pero existen varias alternativas equivalentes para hacerse una idea de la variabilidad de la respuesta explicada por el modelo. Las tres más populares son el pseudo $R^2$ de McFadden (que será el que utilicemos nosotros), el $R^2$ de Cox y Snell (que por construcción no puede alcanzar el $1$) y el $R^2$ de Nagelkerke (una corrección del de Cox y Snell). Vamos a extraer el coeficiente de determinación $R^2$ de McFadden a través de una forma únicamente:

+ Cálculo manual del coeficiente de determinación de McFadden a partir de la siguiente ecuación formal:

\begin{align*}
R^{2}=1 - \frac{Log-Likelihood_1}{Log-Likelihood_0}
\end{align*}

siendo $Log-Likelihood_1$ el logaritmo de la razón de verosimilitud del modelo de regresión logístico con todas las variables y $Log-Likelihood_0$ el logaritmo de la razón de verosimilitud del modelo de regresión logístico constante.

```r
# Estimamos el modelo de regresión logística constante
null = glm(low ~ 1, family = "binomial", data = wdata)
```

```r
# Estimamos el coeficiente de determinación R2 de McFadden
1 - logLik(modelo_logi) / logLik(null)
```

**<u>Interpretación resultados</u>.-** Los valores pueden variar entre $0$ (indicando que las variables independientes no son buenas explicando la probabilidad de la variable respuesta) y $1$ (indicando que el modelo explica bien la probabilidad de la variable respuesta). En este caso, obtenemos un valor $R^2$ de McFadden igual a $0.02548803$ lo que indica que considerando solamente la variable `lwt` no conseguimos mucha capacidad explicativa de la probabilidad de que el peso de un recién nacido sea inferior a $2.5$kg. Luego, si bien, el modelo mejora al cincorporar dicha variable, frente al modelo nulo o constante, sería recomendable incorporar más variables explicativas para mejorar los resultados.

#### 4.8 ¿Cómo extraer la desviación de los residuos y del modelo nulo?

#### 4.9 ¿Cómo extraer el coeficiente de determinación $R^2$ de McFadden?

Recordar que en el modelo de regresión logística no tiene sentido calcular el coeficiente de determinación lineal, $R^2$, ya que estamos modelizando dicha relación a través de la ecuación logística, pero existen varias alternativas equivalentes para hacerse una idea de la variabilidad de la respuesta explicada por el modelo. Las tres más populares son el pseudo $R^2$ de McFadden (que será el que utilicemos nosotros), el $R^2$ de Cox y Snell (que por construcción no puede alcanzar el $1$) y el $R^2$ de Nagelkerke (una corrección del de Cox y Snell). Vamos a extraer el coeficiente de determinación $R^2$ de McFadden a través de una forma únicamente:

+ Cálculo manual del coeficiente de determinación de McFadden a partir de la siguiente ecuación formal:

\begin{align*}
R^{2}=1 - \frac{Log-Likelihood_1}{Log-Likelihood_0}
\end{align*}

siendo $Log-Likelihood_1$ el logaritmo de la razón de verosimilitud del modelo de regresión logístico con todas las variables y $Log-Likelihood_0$ el logaritmo de la razón de verosimilitud del modelo de regresión logístico constante.

```r
# Estimamos el modelo de regresión logística constante
null = glm(Survived ~ 1, family = "binomial", data = wdata)
```

```r
# Estimamos el coeficiente de determinación R2 de McFadden
1 - logLik(modelo_logi) / logLik(null)
```
